```python
import numpy as np
import pandas as pd
```

#Loading Dataset

```python
df = pd.read_csv('Online retail.csv')
df.head()
```

#Data Preprocessing and converting the data into correct form

```python
df_transactions = df['shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil'].str.split(',', expand=True)
df_transactions = df_transactions.apply(lambda x: pd.Series(x.dropna().unique()), axis=1).fillna('')
```

```python
df_transactions.head(10)
```

```python
from mlxtend.preprocessing import TransactionEncoder
transactions_list = df_transactions.apply(lambda row: row[row != ''].tolist(), axis=1).tolist()

te = TransactionEncoder()
te_ary = te.fit(transactions_list).transform(transactions_list)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df_encoded
```

#Implementing Apriori algorithm and association rule mining technique

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
rules = rules.sort_values('lift', ascending=False)
rules
```

```python
rules1 = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
rules1 = rules1.sort_values('lift', ascending=False)
rules1
```

#Analysing the interesting patterns and relationships between the products

```python
for index, row in rules1.iterrows():
  antecedents = list(row['antecedents'])
  consequents = list(row['consequents'])
  lift = row['lift']
  confidence = row['confidence']

  print(f"\nIf a customer buys {antecedents}, they are likely to also buy {consequents}.")
  print(f"Lift: {lift}, Confidence: {confidence}")

  # Provide insights based on the rule
  if 'mineral water' in antecedents and 'green tea' in consequents:
    print("Insight: Customers who buy mineral water often also buy green tea, suggesting a potential healthy lifestyle association.")
  elif 'shrimp' in antecedents and 'frozen smoothie' in consequents:
    print("Insight: Customers who buy shrimp often buy frozen smoothies, indicating a potential interest in convenient and healthy meal options.")
```

**1.What is lift and why is it important in Association rules?**

Lift is an important metric in association rule mining, and it measures how much more likely the consequent is to occur when the antecedent is present, compared to when the antecedent is absent.

Importance:

-> Lift allows us to see if an association rule is simply due to chance or if there is a real relationship between items. While confidence tells the reliability of a rule, lift helps you understand whether the rule is meaningful.

-> If the lift value is high (greater than 1), it implies that the items are purchased together more frequently than by chance, making it a valuable insight for things like recommendation systems or promotional bundling.


**2.What is support and Confidence. How do you calculate them?**

Support and Confidence are two fundamental metrics used in association rule mining to evaluate the strength and significance of a rule. They help in identifying relationships between items in a transaction dataset.

**-> Support:**

Support measures how frequently a particular item or itemset appears in the dataset. It reflects the proportion of transactions that contain the itemset out of the total number of transactions. In simple terms, it tells you how popular an item or itemset is.

Support(A) = Number of transactions containing A / Total number of transactions

**-> Confidence:**

Confidence measures how often the rule {A} ⇒ {B} is found to be true. It indicates the likelihood that a transaction containing item A will also contain item B. In other words, it is the conditional probability that B is purchased given that A is purchased.

Confidence(A->B) = Support(A,B) / Support(A)
