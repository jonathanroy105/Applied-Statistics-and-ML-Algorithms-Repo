```python
df1 = df.copy()
```

**Stanadardization**

Standardization is a scaling method where the values are centered around mean with a unit standard deviation.
It means if we calculate mean and standard deviation of standard scores it will be 0 and 1 respectively.

The formula is given by z=x-mu/sigma,

where mu is the mean and sigma is the standard deviation of the given distribution.

Z is the standard score and it represents the number of standard deviations above or below the mean that a specific observation falls.

```python
for col in df.select_dtypes(include=['float64','int64']).columns:
  mean = df[col].mean()
  std = df[col].std()
  df1[col] = (df[col] - mean) / std
```

```python
num_cols = df.select_dtypes(include=['float64','int64']).columns
```

```python
for column in num_cols:
    plt.figure(figsize=(11, 5))

    #original distribution
    plt.subplot(1, 2, 1)
    plt.hist(df[column], bins=30, color='blue', alpha=0.7)
    plt.title(f'Original Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    #standardized distribution
    plt.subplot(1, 2, 2)
    plt.hist(df1[column], bins=30, color='green', alpha=0.7)
    plt.title(f'Standardized Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```

**Converting categorical data into dummy variables (one-hot encoding) is essential because:**

1.Many ML algorithms need numerical input, and categorical data must be converted to numbers.

2.It prevents the model from incorrectly assuming an order or hierarchy among categories.

3.It ensures each category is treated independently, allowing the model to learn accurate relationships.

4.One-hot encoding makes categorical data usable across various algorithms, especially those that require numeric input.

**Data preprocessing steps like standardization and one-hot encoding are crucial in data analysis and machine learning for several reasons:**

**1.Standardization:**

**Ensures Uniform Scale:** Standardization transforms numerical features to have a mean of 0 and a standard deviation of 1.
This helps in normalizing the scale of features, which is important for algorithms sensitive to the scale of data,
such as gradient descent-based methods.

**Improves Convergence:** For algorithms that rely on distance metrics, standardization ensures that features contribute equally to the distance calculations,
improving the algorithms performance and convergence speed.

**Reduces Bias:** By scaling features to a common range, standardization prevents features with larger scales from dominating the learning process.


**2.One-Hot Encoding:**

**Transforms Categorical Data:** Many machine learning algorithms require numerical input.
One-hot encoding converts categorical variables into binary vectors, allowing models to process categorical data effectively.

**Avoids Ordinal Assumptions:** Directly encoding categories as integers might imply an ordinal relationship.
One-hot encoding avoids this issue by representing each category as a separate binary feature, ensuring no false ordering.

**Ensures Model Compatibility:** One-hot encoding makes categorical features compatible with algorithms that cannot handle non-numeric data,
ensuring that all features are usable in model training.
