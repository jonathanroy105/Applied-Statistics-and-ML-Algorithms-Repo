```python
x = df.drop('Type', axis=1)
y = df['Type']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled[:5]
```

#Performing dimensionality reduction using PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
x_pca = pca.fit_transform(x_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Scree Plot - Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
explained_variance[:10], cumulative_variance[:10]
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(x_scaled)
```

```python
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title('K-means Clustering')
plt.xlabel('Alcohol')
plt.ylabel('Malic')
plt.show()
```

```python
plt.scatter(x_scaled[:,5], x_scaled[:,6], c=clusters, cmap='viridis', marker='o')
plt.title('K-means Clustering')
plt.xlabel('Phenols')
plt.ylabel('Flavanoids')
plt.show()
```

```python
sil_score = silhouette_score(x_scaled, clusters)
db_score = davies_bouldin_score(x_scaled, clusters)
```

```python
sil_score, db_score
```

```python
pca = PCA(n_components=2)
x_pca_transformed = pca.fit_transform(x_scaled)
```

```python
kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(x_pca_transformed)
```

```python
plt.scatter(x_pca_transformed[:, 0], x_pca_transformed[:, 1], c=clusters_pca, cmap='viridis', marker='o')
plt.title('K-means Clustering on PCA-Transformed Data')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```

```python
sil_score_pca = silhouette_score(x_pca_transformed, clusters_pca)
db_score_pca = davies_bouldin_score(x_pca_transformed, clusters_pca)
```

```python
sil_score_pca, db_score_pca
```

#Comparision and Analysis

**1.Comparing the clustering results obtained from the original dataset and PCA-transformed data.**

The primary way to compare results is through evaluation metrics like:

**Silhouette Score:** Measures how similar an object is to its own cluster compared to other clusters.

**Davies–Bouldin Index:** Measures how well the clusters are separated (lower values indicate better clustering).


The Silhouette score is higher and the Davies-Bouldin index are almost same for the PCA-transformed data, which means that clustering performance is better on the PCA-transformed data compared to the original data.

Dimensionality reduction using PCA in this case has enhanced the clustering by removing noise and redundant features, leading to more distinct and well-separated clusters.

**2.Similarities or differences observed in the clustering results.**

-> Both approaches aim to group the wine data into 3 clusters.

-> The general arrangement of clusters is somewhat similar visually.

-> Clusters in PCA-transformed data might appear tighter and more separated.

-> Clusters formed on PCA-transformed data are interpreted based on principal components, which are combinations of original features. In contrast, clusters from the original data are directly based on the original features.


**3.Impact of dimensionality reduction technique on clustering performance.**

-> Processing fewer features requires less computational resources, leading to faster execution of clustering algorithms, especially for large datasets.

-> By focusing on the principal components that capture the most variance, dimensionality reduction can filter out noise and less informative features, potentially leading to more accurate cluster assignments.

->  Reducing the number of dimensions can make the clusters more distinct and easier to separate, improving the performance of clustering algorithms.

**Clustering directly on the original dataset:**

-> CLusters are based on original features, making them easier to understand.

-> Can lead to the curse of dimensionality, impacting distance calculations and clustering accuracy.


**Clustering on PCA-transformed data:**

->  Improves efficiency and potentially enhances clustering due to noise reduction.

-> Discarding less important principal components might result in some information loss.

#Conclusions and Insights

**Key findings and insights from this assignments:**

-> The wine dataset was successfully clustered into three distinct groups using K-means clustering and both original data and PCA-transformed data were used for clustering.

-> PCA significantly reduced the dimensionality of the dataset while retaining a majority of the variance. Clustering on PCA-transformed data resulted in potentially better performance, as indicated by the silhouette score. Dimensionality reduction using PCA can enhance clustering by reducing noise and improving cluster separation.

-> Silhouette score and Davies-Bouldin index were employed to evaluate clustering performance. PCA-transformed data yielded a slightly higher silhouette score, suggesting improved cluster quality.

-> Clustering on original data allows for direct interpretation using original features but can suffer from the curse of dimensionality.
Clustering on PCA-transformed data provides more efficient clustering, potentially improving accuracy but requires interpreting clusters based on principal components.


**Recommendations for using original data vs PCA-transformed data for clustering:**

1.Use original data for clustering:

-> When you need to understand the clusters based on the original features and their relationships, clustering on the original data is preferred.

-> For smaller datasets with a manageable number of features, clustering directly on the original data might be sufficient and easier to implement.


2.Use PCA transformed data for clustering.

-> If the dataset is large and has a high number of features, PCA-transformed data can significantly reduce computational costs and speed up the clustering process.

-> PCA helps remove less informative features and noise, which can lead to more accurate and well-defined clusters.
